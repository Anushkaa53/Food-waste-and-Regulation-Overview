```python
#STEP 1 - Loading the Dataset
import pandas as pd

# Load the dataset
file_path = '/Users/annushkaa/Downloads/FAOLEX_Food.csv'
df = pd.read_csv(file_path)

# Inspect the first few rows
df.head()

```

```python
#STEP 2 - Handling Missing data
```

```python
# Check for missing values
df.isnull().sum()

# Drop rows with missing essential values (if applicable)
df_cleaned = df.dropna(subset=['Country/Territory', 'Date of original text'])

# Optionally, fill missing values for non-critical fields
df_cleaned['Last amended date'].fillna('Not Amended', inplace=True)

df_cleaned.head()
```

```python
#STEP 3 - Parsing the dates
```

```python
# Convert date columns to datetime
df_cleaned['Date of original text'] = pd.to_datetime(df_cleaned['Date of original text'], errors='coerce')
df_cleaned['Last amended date'] = pd.to_datetime(df_cleaned['Last amended date'], errors='coerce')

# Inspect the cleaned data
df_cleaned.info()
```

```python
#STEP 4 - Extract Additional Information
# Extract year of regulation
df_cleaned['Year'] = df_cleaned['Date of original text'].dt.year

# Calculate the time difference between original and last amendment (if applicable)
df_cleaned['Amendment Lag (Years)'] = (df_cleaned['Last amended date'] - df_cleaned['Date of original text']).dt.days // 365
```

```python
#STEP 5 - Export Cleaned Data for Tableau

# Define the path where you want to save the CSV file
file_path = '/Users/annushkaa/Downloads/Cleaned_Food_Regulations.csv'

# Save the DataFrame to a CSV file
df.to_csv(file_path, index=False)
```
